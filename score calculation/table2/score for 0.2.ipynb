```python

import pandas as pd

import numpy as np

def compute_relative_scores(results: dict, metric: str, eps: float = 1e-12):
    algs = list(results.keys())
    ks = sorted(next(iter(results.values())).keys())

    Y = np.array([[results[a][k][metric] for a in algs] for k in ks], dtype=float)

    worst = Y.max(axis=1)
    best = Y.min(axis=1)
    denom = worst - best

    R = np.empty_like(Y)
    mask = denom > eps
    if mask.any():
        R[mask, :] = (worst[mask, None] - Y[mask, :]) / (denom[mask, None])
    if (~mask).any():
        R[~mask, :] = 1.0

    R = np.clip(R, 0.0, 1.0)

    r_df = pd.DataFrame(R, index=ks, columns=algs)
    p = r_df.mean(axis=0).reindex(algs)

    return r_df, p


results = {
    "SRBF": {
        1: {"distance": 0.4788, "time": 0.33},
        2: {"distance": 1.4562, "time": 0.36},
        3: {"distance": 0.5055, "time": 0.34},
        4: {"distance": 0.4699, "time": 0.30},
        5: {"distance": 0.4637, "time": 0.37}
    },
    "NDCT_standard_PSO": {
        1: {"distance": 2.5122, "time": 0.14},
        2: {"distance": 2.4318, "time": 0.13},
        3: {"distance": 1.5477, "time": 0.15},
        4: {"distance": 1.4590, "time": 0.12},
        5: {"distance": 1.4951, "time": 0.13}
    },
    "NDCT_modified_PSO": {
        1: {"distance": 0.4724, "time": 0.21},
        2: {"distance": 0.3978, "time": 0.23},
        3: {"distance": 0.5589, "time": 0.20},
        4: {"distance": 0.3891, "time": 0.22},
        5: {"distance": 0.3640, "time": 0.24}
    },
    "DYCORS": {
        1: {"distance": 0.4274, "time": 0.46},
        2: {"distance": 0.3529, "time": 0.41},
        3: {"distance": 0.7291, "time": 0.48},
        4: {"distance": 0.3142, "time": 0.43},
        5: {"distance": 0.3457, "time": 0.44}
    },
    "COBYLA": {
        1: {"distance": 4.7841, "time": 0.01},
        2: {"distance": 2.8199, "time": 0.01},
        3: {"distance": 3.7555, "time": 0.01},
        4: {"distance": 5.7933, "time": 0.01},
        5: {"distance": 3.7680, "time": 0.01}
    },
    "CUATRO": {
        1: {"distance": 4.4621, "time": 3.95},
        2: {"distance": 4.4833, "time": 3.91},
        3: {"distance": 4.4177, "time": 3.96},
        4: {"distance": 4.4529, "time": 3.87},
        5: {"distance": 4.4766, "time": 3.93}
    },
    "COBYQA": {
        1: {"distance": 5.9271, "time": 0.42},
        2: {"distance": 5.9022, "time": 0.44},
        3: {"distance": 5.9612, "time": 0.40},
        4: {"distance": 5.9339, "time": 0.43},
        5: {"distance": 5.9455, "time": 0.39}
    },
    "ENTMOOT": {
        1: {"distance": 1.9721, "time": 37.60},
        2: {"distance": 1.0156, "time": 38.12},
        3: {"distance": 1.9512, "time": 37.25},
        4: {"distance": 1.9898, "time": 37.95},
        5: {"distance": 1.9634, "time": 38.33}
    }
}



r_time, p_time = compute_relative_scores(results, metric="time")
r_dist, p_dist = compute_relative_scores(results, metric="distance")


df_summary = pd.DataFrame({
    "Time Score (mean)": p_time.round(3),
    "Distance Score (mean)": p_dist.round(3)
})
print(df_summary.to_markdown())

for val in df_summary["Time Score (mean)"]:
    print(val)

for val in df_summary["Distance Score (mean)"]:
    print(val)
```
