```python

import pandas as pd

import numpy as np

def compute_relative_scores(results: dict, metric: str, eps: float = 1e-12):
    algs = list(results.keys())
    ks = sorted(next(iter(results.values())).keys())

    Y = np.array([[results[a][k][metric] for a in algs] for k in ks], dtype=float)

    worst = Y.max(axis=1)
    best = Y.min(axis=1)
    denom = worst - best

    R = np.empty_like(Y)
    mask = denom > eps
    if mask.any():
        R[mask, :] = (worst[mask, None] - Y[mask, :]) / (denom[mask, None])
    if (~mask).any():
        R[~mask, :] = 1.0

    R = np.clip(R, 0.0, 1.0)

    r_df = pd.DataFrame(R, index=ks, columns=algs)
    p = r_df.mean(axis=0).reindex(algs)

    return r_df, p


results = {
    "SRBF": {
        1: {"distance": 0.5923, "time": 0.35},
        2: {"distance": 0.4587, "time": 0.33},
        3: {"distance": 0.4719, "time": 0.32},
        4: {"distance": 0.5076, "time": 0.37},
        5: {"distance": 0.4398, "time": 0.31}
    },
    "NDCT_standard_PSO": {
        1: {"distance": 1.8321, "time": 0.12},
        2: {"distance": 1.2746, "time": 0.13},
        3: {"distance": 0.7059, "time": 0.10},
        4: {"distance": 0.7922, "time": 0.18},
        5: {"distance": 1.3487, "time": 0.14}
    },
    "NDCT_modified_PSO": {
        1: {"distance": 0.5395, "time": 0.18},
        2: {"distance": 0.4147, "time": 0.20},
        3: {"distance": 0.5231, "time": 0.15},
        4: {"distance": 0.2879, "time": 0.21},
        5: {"distance": 0.5612, "time": 0.19}
    },
    "DYCORS": {
        1: {"distance": 0.5341, "time": 0.37},
        2: {"distance": 0.8815, "time": 0.38},
        3: {"distance": 0.5912, "time": 0.34},
        4: {"distance": 0.7358, "time": 0.40},
        5: {"distance": 0.3026, "time": 0.36}
    },
    "COBYLA": {
        1: {"distance": 5.7511, "time": 0.01},
        2: {"distance": 5.8259, "time": 0.01},
        3: {"distance": 5.7894, "time": 0.01},
        4: {"distance": 5.7042, "time": 0.01},
        5: {"distance": 5.8320, "time": 0.01}
    },
    "CUATRO": {
        1: {"distance": 3.4529, "time": 4.29},
        2: {"distance": 3.5180, "time": 4.18},
        3: {"distance": 3.3965, "time": 4.44},
        4: {"distance": 3.4431, "time": 4.11},
        5: {"distance": 3.5098, "time": 4.38}
    },
    "COBYQA": {
        1: {"distance": 6.0998, "time": 0.47},
        2: {"distance": 6.0721, "time": 0.52},
        3: {"distance": 6.0415, "time": 0.59},
        4: {"distance": 6.0152, "time": 0.44},
        5: {"distance": 6.0836, "time": 0.61}
    },
    "ENTMOOT": {
        1: {"distance": 3.1420, "time": 39.92},
        2: {"distance": 3.0187, "time": 41.18},
        3: {"distance": 3.1055, "time": 38.27},
        4: {"distance": 3.0711, "time": 40.45},
        5: {"distance": 3.0873, "time": 37.66}
    }
}




r_time, p_time = compute_relative_scores(results, metric="time")
r_dist, p_dist = compute_relative_scores(results, metric="distance")


df_summary = pd.DataFrame({
    "Time Score (mean)": p_time.round(3),
    "Distance Score (mean)": p_dist.round(3)
})
print(df_summary.to_markdown())

for val in df_summary["Time Score (mean)"]:
    print(val)


for val in df_summary["Distance Score (mean)"]:
    print(val)
```
