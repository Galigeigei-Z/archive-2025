```python

import pandas as pd

import numpy as np

def compute_relative_scores(results: dict, metric: str, eps: float = 1e-12):
    algs = list(results.keys())
    ks = sorted(next(iter(results.values())).keys())

    Y = np.array([[results[a][k][metric] for a in algs] for k in ks], dtype=float)

    worst = Y.max(axis=1)
    best = Y.min(axis=1)
    denom = worst - best

    R = np.empty_like(Y)
    mask = denom > eps
    if mask.any():
        R[mask, :] = (worst[mask, None] - Y[mask, :]) / (denom[mask, None])
    if (~mask).any():
        R[~mask, :] = 1.0

    R = np.clip(R, 0.0, 1.0)

    r_df = pd.DataFrame(R, index=ks, columns=algs)
    p = r_df.mean(axis=0).reindex(algs)

    return r_df, p


results = {
    "SRBF": {
        1: {"distance": 0.4497, "time": 0.34},
        2: {"distance": 0.5017, "time": 0.32},
        3: {"distance": 0.4725, "time": 0.33},
        4: {"distance": 0.4889, "time": 0.31},
        5: {"distance": 0.4603, "time": 0.36}
    },
    "NDCT_standard_PSO": {
        1: {"distance": 0.7189, "time": 0.10},
        2: {"distance": 1.3561, "time": 0.12},
        3: {"distance": 0.7952, "time": 0.11},
        4: {"distance": 1.2805, "time": 0.14},
        5: {"distance": 0.8410, "time": 0.19}
    },
    "NDCT_modified_PSO": {
        1: {"distance": 0.5228, "time": 0.19},
        2: {"distance": 0.3018, "time": 0.22},
        3: {"distance": 0.4984, "time": 0.14},
        4: {"distance": 0.3267, "time": 0.21},
        5: {"distance": 0.5451, "time": 0.19}
    },
    "DYCORS": {
        1: {"distance": 0.9215, "time": 0.36},
        2: {"distance": 0.7001, "time": 0.36},
        3: {"distance": 0.8742, "time": 0.38},
        4: {"distance": 0.7420, "time": 0.35},
        5: {"distance": 0.9105, "time": 0.39}
    },
    "COBYLA": {
        1: {"distance": 5.7845, "time": 0.01},
        2: {"distance": 5.7343, "time": 0.01},
        3: {"distance": 5.8210, "time": 0.01},
        4: {"distance": 5.7655, "time": 0.01},
        5: {"distance": 5.6999, "time": 0.01}
    },
    "CUATRO": {
        1: {"distance": 3.4406, "time": 4.18},
        2: {"distance": 3.4406, "time": 4.51},
        3: {"distance": 3.3899, "time": 4.32},
        4: {"distance": 3.5211, "time": 4.25},
        5: {"distance": 3.4602, "time": 4.41}
    },
    "COBYQA": {
        1: {"distance": 6.0604, "time": 0.63},
        2: {"distance": 6.0604, "time": 0.43},
        3: {"distance": 6.1120, "time": 0.58},
        4: {"distance": 6.0044, "time": 0.52},
        5: {"distance": 6.0815, "time": 0.49}
    },
    "ENTMOOT": {
        1: {"distance": 3.0865, "time": 39.41},
        2: {"distance": 3.0865, "time": 38.09},
        3: {"distance": 3.0210, "time": 40.12},
        4: {"distance": 3.1208, "time": 37.55},
        5: {"distance": 3.0642, "time": 41.33}
    }
}



r_time, p_time = compute_relative_scores(results, metric="time")
r_dist, p_dist = compute_relative_scores(results, metric="distance")


df_summary = pd.DataFrame({
    "Time Score (mean)": p_time.round(3),
    "Distance Score (mean)": p_dist.round(3)
})
print(df_summary.to_markdown())

for val in df_summary["Time Score (mean)"]:
    print(val)


for val in df_summary["Distance Score (mean)"]:
    print(val)
```
