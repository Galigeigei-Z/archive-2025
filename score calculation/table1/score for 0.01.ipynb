```python

import pandas as pd

import numpy as np

def compute_relative_scores(results: dict, metric: str, eps: float = 1e-12):
    algs = list(results.keys())
    ks = sorted(next(iter(results.values())).keys())

    Y = np.array([[results[a][k][metric] for a in algs] for k in ks], dtype=float)

    worst = Y.max(axis=1)
    best = Y.min(axis=1)
    denom = worst - best

    R = np.empty_like(Y)
    mask = denom > eps
    if mask.any():
        R[mask, :] = (worst[mask, None] - Y[mask, :]) / (denom[mask, None])
    if (~mask).any():
        R[~mask, :] = 1.0

    R = np.clip(R, 0.0, 1.0)

    r_df = pd.DataFrame(R, index=ks, columns=algs)
    p = r_df.mean(axis=0).reindex(algs)

    return r_df, p


results = {
    "SRBF": {
        1: {"distance": 1.1530, "time": 0.32},
        2: {"distance": 0.2899, "time": 0.32},
        3: {"distance": 0.2139, "time": 0.30},
        4: {"distance": 1.2907, "time": 0.32},
        5: {"distance": 1.4142, "time": 0.32}
    },
    "PSO": {
        1: {"distance": 1.5369, "time": 0.13},
        2: {"distance": 0.9728, "time": 0.10},
        3: {"distance": 1.2885, "time": 0.12},
        4: {"distance": 0.5584, "time": 0.12},
        5: {"distance": 1.9111, "time": 0.10}
    },
    "mPSO": {
        1: {"distance": 0.4295, "time": 0.20},
        2: {"distance": 0.2072, "time": 0.19},
        3: {"distance": 0.3995, "time": 0.18},
        4: {"distance": 1.2101, "time": 0.19},
        5: {"distance": 0.9446, "time": 0.18}
    },
    "DYCORS": {
        1: {"distance": 0.8463, "time": 0.36},
        2: {"distance": 0.4462, "time": 0.37},
        3: {"distance": 0.1106, "time": 0.37},
        4: {"distance": 1.4142, "time": 0.54},
        5: {"distance": 0.9499, "time": 0.58}
    },
    "COBYLA": {
        1: {"distance": 5.8482, "time": 0.01},
        2: {"distance": 5.8250, "time": 0.01},
        3: {"distance": 5.8429, "time": 0.01},
        4: {"distance": 5.0023, "time": 0.00},
        5: {"distance": 5.0023, "time": 0.01}
    },
    "CUATRO": {
        1: {"distance": 3.4406, "time": 5.44},
        2: {"distance": 3.4406, "time": 5.32},
        3: {"distance": 3.4406, "time": 5.50},
        4: {"distance": 4.1307, "time": 3.90},
        5: {"distance": 4.1002, "time": 3.95}
    },
    "COBYQA": {
        1: {"distance": 6.0644, "time": 0.48},
        2: {"distance": 6.0644, "time": 0.48},
        3: {"distance": 6.0644, "time": 0.50},
        4: {"distance": 3.4293, "time": 0.42},
        5: {"distance": 3.4293, "time": 0.51}
    },
    "ENTMOOT": {
        1: {"distance": 2.9850, "time": 38.24},
        2: {"distance": 2.9850, "time": 39.64},
        3: {"distance": 2.9850, "time": 38.61},
        4: {"distance": 1.4142, "time": 31.69},
        5: {"distance": 1.4142, "time": 31.12}
    }
}


r_time, p_time = compute_relative_scores(results, metric="time")
r_dist, p_dist = compute_relative_scores(results, metric="distance")


df_summary = pd.DataFrame({
    "Time Score (mean)": p_time.round(3),
    "Distance Score (mean)": p_dist.round(3)
})
print(df_summary.to_markdown())

for val in df_summary["Time Score (mean)"]:
    print(val)


for val in df_summary["Distance Score (mean)"]:
    print(val)
```
