```python

import pandas as pd

import numpy as np

def compute_relative_scores(results: dict, metric: str, eps: float = 1e-12):
    algs = list(results.keys())
    ks = sorted(next(iter(results.values())).keys())

    Y = np.array([[results[a][k][metric] for a in algs] for k in ks], dtype=float)

    worst = Y.max(axis=1)
    best = Y.min(axis=1)
    denom = worst - best

    R = np.empty_like(Y)
    mask = denom > eps
    if mask.any():
        R[mask, :] = (worst[mask, None] - Y[mask, :]) / (denom[mask, None])
    if (~mask).any():
        R[~mask, :] = 1.0

    R = np.clip(R, 0.0, 1.0)

    r_df = pd.DataFrame(R, index=ks, columns=algs)
    p = r_df.mean(axis=0).reindex(algs)

    return r_df, p


results = {
    "SRBF": {
        1: {"distance": 0.7488, "time": 0.31},
        2: {"distance": 0.7316, "time": 0.31},
        3: {"distance": 0.6952, "time": 0.33},
        4: {"distance": 0.5821, "time": 0.32},
        5: {"distance": 0.6427, "time": 0.34}
    },
    "NDCT_standard_PSO": {
        1: {"distance": 1.5860, "time": 0.12},
        2: {"distance": 0.6821, "time": 0.11},
        3: {"distance": 1.2458, "time": 0.13},
        4: {"distance": 0.9142, "time": 0.12},
        5: {"distance": 1.1036, "time": 0.14}
    },
    "NDCT_modified_PSO": {
        1: {"distance": 1.6247, "time": 0.19},
        2: {"distance": 0.1796, "time": 0.19},
        3: {"distance": 0.4328, "time": 0.21},
        4: {"distance": 0.2871, "time": 0.20},
        5: {"distance": 0.3549, "time": 0.22}
    },
    "DYCORS": {
        1: {"distance": 0.5623, "time": 0.46},
        2: {"distance": 0.3929, "time": 0.48},
        3: {"distance": 0.4157, "time": 0.47},
        4: {"distance": 0.6314, "time": 0.49},
        5: {"distance": 0.4768, "time": 0.45}
    },
    "COBYLA": {
        1: {"distance": 5.7677, "time": 0.01},
        2: {"distance": 5.7796, "time": 0.01},
        3: {"distance": 5.8012, "time": 0.01},
        4: {"distance": 5.7654, "time": 0.01},
        5: {"distance": 5.7839, "time": 0.01}
    },
    "CUATRO": {
        1: {"distance": 3.4406, "time": 4.31},
        2: {"distance": 5.9742, "time": 3.54},
        3: {"distance": 4.1289, "time": 4.12},
        4: {"distance": 3.9321, "time": 3.87},
        5: {"distance": 5.2157, "time": 4.05}
    },
    "COBYQA": {
        1: {"distance": 6.0105, "time": 0.52},
        2: {"distance": 6.0342, "time": 0.46},
        3: {"distance": 5.9874, "time": 0.49},
        4: {"distance": 6.0215, "time": 0.44},
        5: {"distance": 6.0128, "time": 0.47}
    },
    "ENTMOOT": {
        1: {"distance": 3.1770, "time": 39.23},
        2: {"distance": 3.1770, "time": 38.90},
        3: {"distance": 3.2258, "time": 39.10},
        4: {"distance": 3.1437, "time": 38.75},
        5: {"distance": 3.1896, "time": 39.45}
    }
}




r_time, p_time = compute_relative_scores(results, metric="time")
r_dist, p_dist = compute_relative_scores(results, metric="distance")


df_summary = pd.DataFrame({
    "Time Score (mean)": p_time.round(3),
    "Distance Score (mean)": p_dist.round(3)
})
print(df_summary.to_markdown())

for val in df_summary["Time Score (mean)"]:
    print(val)

for val in df_summary["Distance Score (mean)"]:
    print(val)
```
