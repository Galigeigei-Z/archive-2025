```python

import pandas as pd


import numpy as np

def compute_relative_scores(results: dict, metric: str, eps: float = 1e-12):
    algs = list(results.keys())
    ks = sorted(next(iter(results.values())).keys())

    Y = np.array([[results[a][k][metric] for a in algs] for k in ks], dtype=float)

    worst = Y.max(axis=1)
    best = Y.min(axis=1)
    denom = worst - best

    R = np.empty_like(Y)
    mask = denom > eps
    if mask.any():
        R[mask, :] = (worst[mask, None] - Y[mask, :]) / (denom[mask, None])
    if (~mask).any():
        R[~mask, :] = 1.0

    R = np.clip(R, 0.0, 1.0)

    r_df = pd.DataFrame(R, index=ks, columns=algs)
    p = r_df.mean(axis=0).reindex(algs)

    return r_df, p


results = {
    "SRBF": {
        1: {"distance": 0.4011, "time": 0.40},
        2: {"distance": 0.4225, "time": 0.38},
        3: {"distance": 0.3897, "time": 0.44},
        4: {"distance": 0.4152, "time": 0.41},
        5: {"distance": 0.3978, "time": 0.39}
    },
    "NDCT_standard_PSO": {
        1: {"distance": 0.9246, "time": 0.16},
        2: {"distance": 0.8754, "time": 0.15},
        3: {"distance": 0.9520, "time": 0.17},
        4: {"distance": 0.8999, "time": 0.14},
        5: {"distance": 0.9403, "time": 0.18}
    },
    "NDCT_modified_PSO": {
        1: {"distance": 0.6289, "time": 0.25},
        2: {"distance": 0.6022, "time": 0.27},
        3: {"distance": 0.6488, "time": 0.23},
        4: {"distance": 0.6115, "time": 0.24},
        5: {"distance": 0.6359, "time": 0.26}
    },
    "DYCORS": {
        1: {"distance": 0.3658, "time": 0.42},
        2: {"distance": 0.3920, "time": 0.40},
        3: {"distance": 0.3542, "time": 0.46},
        4: {"distance": 0.3785, "time": 0.41},
        5: {"distance": 0.3697, "time": 0.43}
    },
    "COBYLA": {
        1: {"distance": 5.7614, "time": 0.01},
        2: {"distance": 5.8122, "time": 0.01},
        3: {"distance": 5.7025, "time": 0.01},
        4: {"distance": 5.7850, "time": 0.01},
        5: {"distance": 5.7308, "time": 0.01}
    },
    "CUATRO": {
        1: {"distance": 3.4406, "time": 3.92},
        2: {"distance": 3.4877, "time": 4.05},
        3: {"distance": 3.3998, "time": 3.80},
        4: {"distance": 3.5211, "time": 4.12},
        5: {"distance": 3.4650, "time": 3.95}
    },
    "COBYQA": {
        1: {"distance": 5.9811, "time": 0.38},
        2: {"distance": 6.0502, "time": 0.36},
        3: {"distance": 5.9220, "time": 0.41},
        4: {"distance": 6.0044, "time": 0.39},
        5: {"distance": 6.0725, "time": 0.37}
    },
    "ENTMOOT": {
        1: {"distance": 0.7954, "time": 37.78},
        2: {"distance": 0.8410, "time": 38.25},
        3: {"distance": 0.7682, "time": 39.12},
        4: {"distance": 0.8233, "time": 36.95},
        5: {"distance": 0.7817, "time": 38.74}
    }
}


r_time, p_time = compute_relative_scores(results, metric="time")
r_dist, p_dist = compute_relative_scores(results, metric="distance")


df_summary = pd.DataFrame({
    "Time Score (mean)": p_time.round(3),
    "Distance Score (mean)": p_dist.round(3)
})
print(df_summary.to_markdown())

for val in df_summary["Time Score (mean)"]:
    print(val)


for val in df_summary["Distance Score (mean)"]:
    print(val)
```
