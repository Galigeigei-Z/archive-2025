```python

import pandas as pd

import numpy as np

def compute_relative_scores(results: dict, metric: str, eps: float = 1e-12):
    algs = list(results.keys())
    ks = sorted(next(iter(results.values())).keys())

    Y = np.array([[results[a][k][metric] for a in algs] for k in ks], dtype=float)

    worst = Y.max(axis=1)
    best = Y.min(axis=1)
    denom = worst - best

    R = np.empty_like(Y)
    mask = denom > eps
    if mask.any():
        R[mask, :] = (worst[mask, None] - Y[mask, :]) / (denom[mask, None])
    if (~mask).any():
        R[~mask, :] = 1.0

    R = np.clip(R, 0.0, 1.0)

    r_df = pd.DataFrame(R, index=ks, columns=algs)
    p = r_df.mean(axis=0).reindex(algs)

    return r_df, p


results = {
    "SRBF": {
        1: {"distance": 1.0417, "time": 0.30},
        2: {"distance": 1.0619, "time": 0.19},
        3: {"distance": 0.9260, "time": 0.39},
        4: {"distance": 1.6011, "time": 0.31},
        5: {"distance": 1.1033, "time": 0.31}
    },
    "NDCT_standard_PSO": {
        1: {"distance": 1.5132, "time": 0.10},
        2: {"distance": 0.5093, "time": 0.12},
        3: {"distance": 1.1650, "time": 0.16},
        4: {"distance": 2.4337, "time": 0.12},
        5: {"distance": 1.3731, "time": 0.12}
    },
    "NDCT_modified_PSO": {
        1: {"distance": 0.6581, "time": 0.18},
        2: {"distance": 0.2549, "time": 0.18},
        3: {"distance": 0.7406, "time": 0.27},
        4: {"distance": 0.5342, "time": 0.19},
        5: {"distance": 0.1899, "time": 0.18}
    },
    "DYCORS": {
        1: {"distance": 0.1359, "time": 2.35},
        2: {"distance": 0.0400, "time": 2.35},
        3: {"distance": 1.0222, "time": 2.44},
        4: {"distance": 0.1641, "time": 2.34},
        5: {"distance": 0.0255, "time": 2.37}
    },
    "COBYLA": {
        1: {"distance": 5.9582, "time": 0.01},
        2: {"distance": 5.9582, "time": 0.03},
        3: {"distance": 5.9582, "time": 0.01},
        4: {"distance": 5.9582, "time": 0.01},
        5: {"distance": 5.9582, "time": 0.01}
    },
    "CUATRO": {
        1: {"distance": 3.4406, "time": 5.68},
        2: {"distance": 3.4406, "time": 6.25},
        3: {"distance": 3.4406, "time": 6.16},
        4: {"distance": 3.4406, "time": 5.54},
        5: {"distance": 3.4406, "time": 6.55}
    },
    "COBYQA": {
        1: {"distance": 6.0081, "time": 0.47},
        2: {"distance": 6.0081, "time": 0.45},
        3: {"distance": 6.0081, "time": 0.46},
        4: {"distance": 6.0081, "time": 0.48},
        5: {"distance": 6.0081, "time": 0.49}
    },
    "ENTMOOT": {
        1: {"distance": 0.1512, "time": 35.47},
        2: {"distance": 0.1512, "time": 36.61},
        3: {"distance": 0.1512, "time": 39.42},
        4: {"distance": 0.1512, "time": 36.11},
        5: {"distance": 0.1512, "time": 36.06}
    }
}


r_time, p_time = compute_relative_scores(results, metric="time")
r_dist, p_dist = compute_relative_scores(results, metric="distance")


df_summary = pd.DataFrame({
    "Time Score (mean)": p_time.round(3),
    "Distance Score (mean)": p_dist.round(3)
})
print(df_summary.to_markdown())

for val in df_summary["Time Score (mean)"]:
    print(val)


for val in df_summary["Distance Score (mean)"]:
    print(val)
```
