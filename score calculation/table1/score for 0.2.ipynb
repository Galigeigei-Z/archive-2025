```python

import pandas as pd

import numpy as np

def compute_relative_scores(results: dict, metric: str, eps: float = 1e-12):
    algs = list(results.keys())
    ks = sorted(next(iter(results.values())).keys())

    Y = np.array([[results[a][k][metric] for a in algs] for k in ks], dtype=float)

    worst = Y.max(axis=1)
    best = Y.min(axis=1)
    denom = worst - best

    R = np.empty_like(Y)
    mask = denom > eps
    if mask.any():
        R[mask, :] = (worst[mask, None] - Y[mask, :]) / (denom[mask, None])
    if (~mask).any():
        R[~mask, :] = 1.0

    R = np.clip(R, 0.0, 1.0)

    r_df = pd.DataFrame(R, index=ks, columns=algs)
    p = r_df.mean(axis=0).reindex(algs)

    return r_df, p


results = {
    "SRBF": {
        1: {"distance": 0.4841, "time": 0.32},
        2: {"distance": 0.4623, "time": 0.34},
        3: {"distance": 0.4998, "time": 0.33},
        4: {"distance": 0.4712, "time": 0.31},
        5: {"distance": 0.4586, "time": 0.35}
    },
    "NDCT_standard_PSO": {
        1: {"distance": 1.4983, "time": 0.13},
        2: {"distance": 1.4421, "time": 0.12},
        3: {"distance": 1.5210, "time": 0.14},
        4: {"distance": 1.4678, "time": 0.13},
        5: {"distance": 1.4889, "time": 0.12}
    },
    "NDCT_modified_PSO": {
        1: {"distance": 0.3657, "time": 0.20},
        2: {"distance": 0.4015, "time": 0.21},
        3: {"distance": 0.3428, "time": 0.19},
        4: {"distance": 0.3834, "time": 0.20},
        5: {"distance": 0.3722, "time": 0.22}
    },
    "DYCORS": {
        1: {"distance": 0.3217, "time": 0.48},
        2: {"distance": 0.3465, "time": 0.40},
        3: {"distance": 0.3328, "time": 0.49},
        4: {"distance": 0.3189, "time": 0.41},
        5: {"distance": 0.3512, "time": 0.47}
    },
    "COBYLA": {
        1: {"distance": 5.7906, "time": 0.01},
        2: {"distance": 5.8124, "time": 0.01},
        3: {"distance": 5.7655, "time": 0.01},
        4: {"distance": 5.8041, "time": 0.01},
        5: {"distance": 5.7812, "time": 0.01}
    },
    "CUATRO": {
        1: {"distance": 3.4406, "time": 3.87},
        2: {"distance": 3.4895, "time": 3.92},
        3: {"distance": 3.4210, "time": 3.95},
        4: {"distance": 3.4677, "time": 3.89},
        5: {"distance": 3.4523, "time": 3.93}
    },
    "COBYQA": {
        1: {"distance": 5.9332, "time": 0.41},
        2: {"distance": 5.9105, "time": 0.43},
        3: {"distance": 5.9510, "time": 0.39},
        4: {"distance": 5.9278, "time": 0.42},
        5: {"distance": 5.9401, "time": 0.40}
    },
    "ENTMOOT": {
        1: {"distance": 2.9619, "time": 37.58},
        2: {"distance": 3.0123, "time": 37.80},
        3: {"distance": 2.9475, "time": 37.42},
        4: {"distance": 2.9834, "time": 37.66},
        5: {"distance": 2.9701, "time": 37.71}
    }
}


r_time, p_time = compute_relative_scores(results, metric="time")
r_dist, p_dist = compute_relative_scores(results, metric="distance")


df_summary = pd.DataFrame({
    "Time Score (mean)": p_time.round(3),
    "Distance Score (mean)": p_dist.round(3)
})
print(df_summary.to_markdown())

for val in df_summary["Time Score (mean)"]:
    print(val)


for val in df_summary["Distance Score (mean)"]:
    print(val)
```
